```python
import requests
import base64
import pandas as pd

service_user = "SERVICE-33880"
token = "Et4Gy4oUD8DMTW7mAm01c5te4x2QEny70WEb81g5"

credentials = f"{service_user}:{token}"
encoded_credentials = base64.b64encode(credentials.encode()).decode()

headers = {
    "Authorization": f"Basic {encoded_credentials}",
    "Content-Type": "application/json"
}

base_url = "https://api.hibob.com/v1"

# Dictionary of endpoints to test
endpoints = {
    "People Search": {
        "method": "POST",
        "url": f"{base_url}/people/search",
        "payload": {"fields": ["firstName", "surname", "email"]}
    },
    "List People": {
        "method": "GET",
        "url": f"{base_url}/people",
        "payload": None
    },
    "Departments": {
        "method": "GET",
        "url": f"{base_url}/company/departments",
        "payload": None
    },
    "Teams": {
        "method": "GET",
        "url": f"{base_url}/company/teams",
        "payload": None
    },
    "Holidays": {
        "method": "GET",
        "url": f"{base_url}/company/holidays",
        "payload": None
    },
    "Leave Types": {
        "method": "GET",
        "url": f"{base_url}/company/leave-types",
        "payload": None
    },
    "Working Hours": {
        "method": "GET",
        "url": f"{base_url}/company/working-hours",
        "payload": None
    }
}

# Test each endpoint
results = []
for endpoint_name, endpoint_config in endpoints.items():
    try:
        if endpoint_config["method"] == "GET":
            response = requests.get(endpoint_config["url"], headers=headers)
        else:
            response = requests.post(endpoint_config["url"], headers=headers, json=endpoint_config["payload"])
        
        status = "✅ WORKING" if response.status_code == 200 else f"❌ {response.status_code}"
        results.append({
            "Endpoint": endpoint_name,
            "URL": endpoint_config["url"],
            "Status": status,
            "Response": response.text[:200] if response.status_code != 200 else "Success"
        })
        print(f"{endpoint_name}: {status}")
        
    except Exception as e:
        results.append({
            "Endpoint": endpoint_name,
            "URL": endpoint_config["url"],
            "Status": f"❌ Error: {str(e)}",
            "Response": str(e)
        })
        print(f"{endpoint_name}: ❌ Error - {str(e)}")

# Display results
df_results = pd.DataFrame(results)
print("\n" + "="*80)
print(df_results.to_string(index=False))
```
